```python
#Python for Data Science Tasks:  
#1. Learn Python basics: variables, data types, loops, and functions.
#2. Work with lists, dictionaries, and tuples.
#3. Practice importing libraries like pandas, numpy, and matplotlib.
# Importing the Titanic dataset to perform my tasks
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
titanic_dataset = pd.read_csv(f"{yasserh_titanic_dataset}/Titanic-Dataset.csv")
titanic_dataset.head(2)
```

```python
#Variables and Data Types
# Integer
passenger_count = len(titanic_dataset)

# Float
average_fare = titanic_dataset["Fare"].mean()

# String
ship_name = "Titanic"

# Boolean
has_missing_values = titanic_dataset.isnull().values.any()

# Display all
print("Passenger Count:", passenger_count)
print("Average Fare:", average_fare)
print("Ship Name:", ship_name)
print("Any Missing Values:", has_missing_values)
```

```python
# Loops (For Loop Example) & Functions
# Print first 5 passenger names using loop
print("First 5 passenger names:")
for name in titanic_dataset["Name"][:5]:
    print(name)
    
# Function to categorize passengers by age
def age_group(age):
    if pd.isna(age):
        return "Unknown"
    elif age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

titanic_dataset["Age_Group"] = titanic_dataset["Age"].apply(age_group)
titanic_dataset[["Name", "Age", "Age_Group"]].head()
```

```python
# Python Data Structures :  lists, dictionaries, and tuples

# List of unique embarkation ports
embarked_ports = titanic_dataset["Embarked"].dropna().unique().tolist()
print("Embarkation Ports:", embarked_ports)

# Dictionary of class counts
pclass_counts =  dict(titanic_dataset["Pclass"].value_counts())
print("Passenger Class Counts:", pclass_counts)

# List of survival status labels
survived_counts = ["Survived" if s == 1 else "Not Survived" for s in titanic_dataset["Survived"]]
# Tuples containing (Name, Survived)
name_survival_pairs = list(zip(titanic_dataset["Name"][:5], survived_counts[:5]))
print("Name and Survival (first 5):", name_survival_pairs)
```

3. Data Handling with Pandas

```python
# Step 1: Load CSV File using pandas.read_csv()
''' import pandas as pd
titanic_dataset = pd.read_csv(f"{yasserh_titanic_dataset}/Titanic-Dataset.csv") '''
# Step 2: Explore the Data

print("first few rows using head():\n", titanic_dataset.head())

print("Check data info and column types using .info()\n")
titanic_dataset.info()

print("stats using describe(): \n", titanic_dataset.describe())

# Step 3: Handle Missing Values

print("missing values in each column\n", titanic_dataset.isnull().sum())

# Fill missing values in 'Age' column with the mean
titanic_dataset['Age'] = titanic_dataset['Age'].fillna(titanic_dataset['Age'].mean())

# Drop 'Embarked' column only if it exists 
if 'Embarked' in titanic_dataset.columns:
    titanic_dataset.drop(columns=['Embarked'], inplace=True)

# Check again for any missing values
print("missing values left:\n", titanic_dataset.isnull().sum())
```

4. Basic Visualization

```python
#Step 1: Import Plotting Libraries

import matplotlib.pyplot as plt
import seaborn as sns

# Optional: set style for seaborn
sns.set(style="whitegrid")

# 1. Matplotlib Visualizations
# Line Plot – Age of first 50 passengers

plt.plot(titanic_dataset['Age'][:50])
plt.title("Line Plot - Age of First 50 Passengers")
plt.xlabel("Passenger Index")
plt.ylabel("Age")
plt.grid(True)
plt.show()

#Bar Plot – Passenger Class Count

pclass_counts = titanic_dataset['Pclass'].value_counts()

pclass_counts.plot(kind='bar', color='skyblue')
plt.title("Bar Plot - Passenger Class Distribution")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()

#Scatter Plot – Age vs Fare
plt.scatter(titanic_dataset['Age'], titanic_dataset['Fare'], alpha=0.5)
plt.title("Scatter Plot - Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

# 2. Seaborn Visualizations
# Histogram – Age Distribution
sns.histplot(titanic_dataset['Age'], bins=30, kde=True, color='purple')
plt.title("Seaborn Histogram - Age Distribution")
plt.show()

#Boxplot – Age by Passenger Class
sns.boxplot(x='Pclass', y='Age', data=titanic_dataset, palette="Set2")
plt.title("Seaborn Boxplot - Age by Passenger Class")
plt.show()
```
